```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
# Importing Demo Images
img_r = cv2.imread('demo2_images/liquid_frame100.jpg', 0)
img_b = cv2.imread('demo2_images/liquid_frame000.jpg', 0)

# Display the images
plt.subplot(121), plt.imshow(img_r, cmap='gray'), plt.title('Random Image (100th Frame)')
plt.subplot(122), plt.imshow(img_b, cmap='gray'), plt.title('Background Image (0th Frame)')
plt.show()
```

Method 1 : Background Subtraction

```python
# Subtracting the images
img_sub = cv2.subtract(img_b, img_r)

# Thresholding the image, to obtain binary image
ret, img_sub_thresh = cv2.threshold(img_sub, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)
# Print the used threshold value
print('Threshold Value:', ret)

# Find contours in the image
contours, hierarchy = cv2.findContours(img_sub_thresh, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)

# Sort the contours by size in descending order
contours = sorted(contours, key=lambda x: cv2.contourArea(x), reverse=True)
# Print the contour areas
for i in range(len(contours)):
    print('Contour', i, 'Area:', cv2.contourArea(contours[i]))

# Find the area of the whole image
rows, cols = img_sub_thresh.shape
total_area = rows * cols
print('Total Area:', total_area)

# Calculate the percentage of the largest contour area, rounded to 1 decimal point
percentage = round((cv2.contourArea(contours[0]) / total_area) * 100, 1)
print('Percentage:', percentage, '%')

# Display img_r , img_sub, img_sub_thresh side by side using plt
plt.subplot(131), plt.imshow(img_r, cmap='gray'), plt.title('Rand Img')
plt.subplot(132), plt.imshow(img_sub, cmap='gray'), plt.title('Subtracted Img')
plt.subplot(133), plt.imshow(img_sub_thresh, cmap='gray'), plt.title('Thresholded Image')
plt.show()
```

2. Otsu's thresholding method is a technique used to automatically determine the optimal threshold value for binarizing an image based on its histogram. Unlike global thresholding, where a threshold is chosen arbitrarily, Otsu's method analyzes the image's intensity distribution to find the threshold that minimizes the intra-class variance.This method is particularly effective for images with a bimodal histogram, where two distinct peaks represent the foreground and background intensities. The process can be enhanced by pre-filtering the image to reduce noise, resulting in better binarization. The `cv.threshold()` function with the `cv.THRESH_OTSU` flag in OpenCV is used to implement this method.

4. The developed algorithm processes a video to detect and analyze changes between frames. It begins by performing background subtraction, where the first frame is used as the reference background, and subsequent frames are subtracted from it to highlight differences. The subtracted image is then converted to grayscale and binarized using Otsu's thresholding method to segment the regions of change. Contours are detected in the binary image, and their areas are calculated and sorted to identify the largest change region. The algorithm calculates the percentage of the largest contour's area relative to the total image area and prints this, along with the threshold value and contour details. Additionally, it visually displays the original frame, the subtracted frame, and the thresholded binary image in real-time. This approach effectively identifies and quantifies changes in the video, making it suitable for motion or activity analysis.

# **CODE**

```python
import cv2

# Open the video file
video_path = 'demo2_images/liquidVideo.mp4'
cap = cv2.VideoCapture(video_path)
ret, background = cap.read()

while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break
    img_sub = cv2.subtract(background, frame)

    # Convert the subtracted image to grayscale
    img_sub_gray = cv2.cvtColor(img_sub, cv2.COLOR_BGR2GRAY)

    # Thresholding the image, to obtain binary image
    ret, img_sub_thresh = cv2.threshold(img_sub_gray, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)

    # Print the used threshold value
    print('Threshold Value:', ret)

    # Find contours in the image
    contours, hierarchy = cv2.findContours(img_sub_thresh, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)

    # Sort the contours by size in descending order
    contours = sorted(contours, key=lambda x: cv2.contourArea(x), reverse=True)
    # Print the contour areas
    for i in range(len(contours)):
        print('Contour area:', cv2.contourArea(contours[i]))

    # Find the area of the whole image
    rows, cols = img_sub_thresh.shape
    total_area = rows * cols
    print('Total Area:', total_area)

    # Calculate the percentage of the largest contour area, rounded to 1 decimal point
    percentage = round((cv2.contourArea(contours[0]) / total_area) * 100, 1)
    print('Percentage:', percentage, '%')

    # Display the frame
    cv2.imshow('Video', frame)
    cv2.imshow('Sub', img_sub)
    cv2.imshow('Sub Thresh', img_sub_thresh)

    if cv2.waitKey(25) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```
